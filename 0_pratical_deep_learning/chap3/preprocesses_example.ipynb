```python
%matplotlib inline
import matplotlib.pyplot as plt
from PIL import Image
import csv
import masalachai
```

```python
with open('caltech101/images_list.csv') as f:
    reader = csv.reader(f)
    image_files = ['caltech101/'+row[0] for row in reader]
```

```python
im = masalachai.preprocesses.load_image(image_files[1095], target_size=(128,128))
im /= im.max()
im = masalachai.preprocesses.random_shift(im,0.1,0.125)
im = masalachai.preprocesses.random_zoom(im,(0.7,0.9))
plt.imshow(im.transpose(1,2,0))
```

```python
im = masalachai.preprocesses.load_image(image_files[1095], target_size=(128,128))
im /= im.max()
im = masalachai.preprocesses.random_rotation(im,80)
plt.imshow(im.transpose(1,2,0))
```

```python
im = masalachai.preprocesses.load_image(image_files[1095], target_size=(128,128))
im /= im.max()
im = masalachai.preprocesses.random_shear(im,0.8)
plt.imshow(im.transpose(1,2,0))
```

```python
im = masalachai.preprocesses.load_image(image_files[1095], target_size=(128,128))
im /= im.max()
im = masalachai.preprocesses.random_shift(im,0.3,0.3)
plt.imshow(im.transpose(1,2,0))
```
